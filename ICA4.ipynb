```python
import pandas as pd

# ---------------------------------------------------------
# 1. Load the dataset
# ---------------------------------------------------------
# If running in Google Colab, upload the file manually:
# from google.colab import files
# uploaded = files.upload()

from google.colab import files
uploaded = files.upload()

df = pd.read_csv("lego_setsHB.csv")   # <-- replace with your path if needed

# ---------------------------------------------------------
# 1. Use .info() and .describe()
# ---------------------------------------------------------
print("=== DATASET INFO ===")
print(df.info())

print("\n=== DATASET DESCRIPTION ===")
print(df.describe(include='all'))

# ---------------------------------------------------------
# 2. Boolean filtering: star_rating >= 4
# ---------------------------------------------------------
high_rating = df[df["star_rating"] >= 4]
print("\n=== LEGO SETS WITH STAR RATING >= 4 ===")
print(high_rating)

# ---------------------------------------------------------
# 3. Count number of sets with star_rating >= 4
# ---------------------------------------------------------
count_high_rating = high_rating["star_rating"].count()
print("\nNumber of Lego sets with star_rating >= 4:", count_high_rating)

# ---------------------------------------------------------
# 4. Average list price for sets with star_rating >= 4
# ---------------------------------------------------------
avg_price_high_rating = high_rating["list_price"].mean()
print("\nAverage list price (star_rating >= 4):", avg_price_high_rating)
```
